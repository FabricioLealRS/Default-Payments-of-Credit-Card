# 1. Objetivo do trabalho



Objetivo do projeto é prever quais clientes podem entrar em inadimplência nos próximos meses. Mas para isso, precisou-se definir o que é realmente entendido por Inadimplência de Cartão de Crédito?

A inadimplência do cartão de crédito ocorre quando o portador se torna  inadimplente nos pagamentos do seu cartão de crédito. Se esquecer de pagar o cartão de crédito por um ou dois dias não é considerado inadimplência. Uma inadimplência ocorre quando você deixa de pagar o Valor Mínimo Devido no cartão de crédito por alguns dias.

Em especial, quando se fala de inadimplencia, uma empresa deve estar sempre atenta ao prazo limite para que a cobrança seja considerada uma perda, que difere desde 6 meses a 1 anos ou mais, após este periodo, mesmo que haja a cobrança ainda do devido, o valor de face (valor original do boleto) é computado no balanço e DRE como perdas, e o subsequente pagamento é considerado uma receita com pagamento de immpostos financeiros.

Vale lembrar que ao entrar como perda no DRE, empresas do Lucro Real podem abater de suas bases de calculo de imposto, motivo este que o pagamento extemporaneo a este prazo é calculado imposto.

a) **Tratamento do Desbalanceamento**: Observou-se que o conjunto de dados possui um desbalanceamento entre as classes de pagadores e inadimplentes. Para lidar com esse problema, será utilizada a técnica SMOTE (Synthetic Minority Oversampling Technique) do pacote imblearn para criar dados sintéticos da classe minoritária (inadimplentes) e balancear o conjunto de dados.

## Regularização

Tendo em vista que é nitida a ocorrencia de Overfiting na base de dados, é proposta a realização de Regularização, uma técnica para evitar o sobreajuste, que é um problema comum em modelos de aprendizado de máquina. A regularização L1 (também conhecida como Lasso) adiciona uma penalidade aos coeficientes do modelo com base no valor absoluto dos coeficientes. Isso pode levar a alguns coeficientes se tornando exatamente zero, o que pode ajudar na interpretação do modelo e na redução da complexidade.

```python
from sklearn.linear_model import LogisticRegression

# L1 regularização
clf_l1 = LogisticRegression(penalty='l1', solver='saga').fit(X_train, y_train)

```

**F1 Score: 46.38%**

O F1 Score é uma média harmônica entre precisão e recall e fornece uma única métrica que leva em conta ambos. 46.38% é uma pontuação mediana que indica um equilíbrio entre precisão e recall, mas ainda assim está longe de ser ideal.

# Discussão e Conclusão

A análise dos modelos apresentados revela um alto grau de eficácia na classificação. Atingir métricas tão altas em precisão, recall e AUC é raro em muitas aplicações do mundo real e pode levantar algumas questões pertinentes:

**Sobreajuste:** Dada a perfeição de muitas das métricas em vários modelos, é crucial considerar a possibilidade de sobreajuste. Modelos que se ajustam muito bem aos dados de treinamento podem não se sair tão bem com novos dados. Seria útil realizar uma validação cruzada ou usar um conjunto de dados totalmente separado para verificar a consistência do desempenho do modelo.

**Qualidade dos Dados:** Dados limpos e bem preparados podem contribuir para a eficácia do modelo, mas é vital assegurar que os dados refletem cenários do mundo real e não foram de alguma forma simplificados ou polarizados, levando a resultados inflacionados.

**Complexidade do Problema:** Nem todos os problemas de classificação são complexos. Em situações onde as classes são facilmente separáveis com base nas características fornecidas, é possível que modelos diferentes alcancem alta precisão.

**Conclusão**
Os modelos de Machine Learning avaliados neste estudo demonstraram alta eficácia na classificação, com a maioria atingindo métricas perfeitas ou quase perfeitas em treinamento e teste. Embora isso seja encorajador, é essencial abordar esses resultados com cautela. Validação adicional, preferencialmente em novos conjuntos de dados ou em cenários do mundo real, seria benéfica para confirmar a robustez e a generalização destes modelos.

Além disso, enquanto os modelos como o SVC não atingiram a perfeição como os outros, eles ainda apresentaram um desempenho admirável e podem ser preferidos em situações onde é necessário um compromisso entre precisão e complexidade computacional.

# Limitações e sugestão de pesquisas futuras

**Limitações**

**Dados Disponíveis:** As métricas de desempenho excepcionais observadas podem indicar que o conjunto de dados usado não captura toda a variabilidade e complexidade de cenários do mundo real. Isso pode resultar em modelos que têm desempenho excelente nos dados atuais, mas que podem falhar ao serem expostos a novas situações.

**Possibilidade de Sobreajuste:** Dada a perfeição ou quase perfeição de algumas métricas, há sempre a preocupação de que os modelos possam estar se ajustando demais aos dados de treinamento e, portanto, possam não generalizar bem para novos dados.

**Complexidade Computacional:** Modelos como o SVC, embora eficazes, podem ser computacionalmente intensivos, especialmente em conjuntos de dados maiores. Esta complexidade pode tornar tais modelos menos viáveis em aplicações em tempo real ou onde o tempo de resposta é crítico.

**Interpretabilidade do Modelo:** Modelos complexos, como florestas aleatórias e XGBoost, podem ser considerados "caixas-pretas", tornando-se desafiadores de interpretar em comparação com modelos mais simples, como regressão logística.

**Sugestões para Pesquisas Futuras**

**Validação Externa:** Seria benéfico validar os modelos usando um conjunto de dados totalmente separado, idealmente coletado de uma fonte diferente, para avaliar a generalização dos modelos em diferentes cenários.

**Uso de Técnicas de Regularização:** Para abordar a possibilidade de sobreajuste, técnicas de regularização podem ser aplicadas para restringir a complexidade do modelo e melhorar a generalização.

**Exploração de Modelos de Aprendizado Profundo:** Dado o sucesso dos modelos tradicionais, pode ser interessante explorar redes neurais e outras técnicas de aprendizado profundo para ver se elas oferecem alguma vantagem em termos de desempenho ou interpretabilidade.

**Engenharia de Recursos Adicional:** A introdução de novas características ou a transformação das existentes pode oferecer insights adicionais e melhorar ainda mais o desempenho do modelo.

**Análise de Causa Raiz dos Erros:** Para os poucos erros cometidos por modelos como o SVC, uma análise aprofundada desses casos específicos pode oferecer insights sobre as nuances do problema e áreas onde o modelo pode ser melhorado.

**Consideração de Abordagens Baseadas em Ensemble:** Combinar as previsões de vários modelos pode oferecer um desempenho melhorado e uma robustez adicional contra erros específicos de um modelo individual.
